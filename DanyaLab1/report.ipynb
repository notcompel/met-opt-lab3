```python
import numpy as np
import method.gradient_descent as gd
import method.dichotomy_gradient as dg
import method.wolfe_gradient as wg
import plot.plotter as plotter
from random import uniform
```

```python
def f(x):
    #return 10 * x[0] ** 2 + x[1] ** 2
    return 4 * x[0] ** 2 + 9 * x[1] ** 2 - 4 * x[0] * x[1] - 2 * x[0] + 12 * x[1] + 7


def grad(x):
    h = 1e-5
    return (f(x[:, np.newaxis] + h * np.eye(2)) - f(x[:, np.newaxis] - h * np.eye(2))) / (2 * h)
```

# $f(x, y) = 10x^2 + y^2$
# $f(x, y) = 4x^2 + 9y^2 - 4xy - 2x + 12y + 7$

```python
def print_info(name, start, points, grad_calc, func_calc):
    print("start:", start)
    print("{}:".format(name))
    print(points[-1], f(points[-1]))
    print("gradient calculations:", grad_calc)
    print("function calculations:", func_calc)


def plot_from_start(start):
    lr = 0.05
    (points1, grad_calc1, func_calc1) = gd.gradient_descent(f, grad, start, 1e-4, lr, 100000)
    print_info("gradient descent", start, points1, grad_calc1, func_calc1)

    (points2, grad_calc2, func_calc2) = dg.dichotomy_gradient(f, grad, start, 1e-4, 1e-3, 0.1)
    print_info("dichotomy gradient", start, points2, grad_calc2, func_calc2)

    (points3, grad_calc3, func_calc3) = wg.wolfe_gradient(f, grad, start, 1e-4, 1e-4, 0.9, 0.1)
    print_info("wolfe gradient", start, points3, grad_calc3, func_calc3)

    plotter.multiple_points_over_contour(f, points1, points2, points3).show()
```

```python
plot_from_start([-1, 2])
```

```python
lr_data = []
grad_data = []
func_data = []
start = [uniform(-2, 2), uniform(-2, 2)]
for lr in range(1, 1000, 2):
    points1, grad_calc1, _ = gd.gradient_descent(f, grad, start, 1e-4, 1 / lr, 1000)
    lr_data.append(1 / lr)
    grad_data.append(grad_calc1)
    func_data.append(min(f(points1[-1]), 4))
plotter.plot_by_two_coordinates(lr_data, grad_data, "lr", False).show()
plotter.plot_by_two_coordinates(lr_data, func_data, "lr", False, "result value")
```
